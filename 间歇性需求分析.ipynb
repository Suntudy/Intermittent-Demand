# 计算数据集的 ADI 和 CV2
ADI = mean(相邻两次非零需求之间的时间间隔)
1.32 是 Croston 文献里的经典阈值：ADI > 1.32 是间歇性需求。
CV2 = (方差)/(均值的平方) 阈值：0.49
| 类别           | ADI | CV² | 特征               |
| ------------ | --- | --- | ---------------- |
| Smooth       | 低   | 低   | 常卖、量稳定           |
| Erratic      | 低   | 高   | 常卖、量不稳           |
| Intermittent | 高   | 低   | 不常卖、量稳定          |
| Lumpy        | 高   | 高   | **不常卖 + 量乱**（最难） |

# GAN 生成式对抗网络
生成模型+判别模型

判别模型：给定一张图，判断图中的动物是什么类别

生成模型：给一系列猫的图片，生成一张新的猫咪

训练过程中，生成网络G的目标就是尽量生成真实的图片去欺骗判别网络D，而D的目标就是尽量把G生成的图片和真实图片分别开来，这样，G和D构成了一个动态的博弈过程。

生成模型功能：比作是一个样本生成器，输入一个噪声/样本，然后把它包装成一个逼真的样本，也就是输出。生成模型的目标是：使得判别器无法判断，无论对于真假样本，输出结果概率都是0.5

判别模型：比作一个二分类器（如同01分类器），来判断输入的样本是真是假。（输出值大于0.5还是小于0.5）

GAN的优势在于能自动学习原始真实样本集的数据分布，不管这个分布多么复杂，只要训练的足够好就可以学出来。而不是传统的机器学习，需要提前知道数据属于高斯分布，但不知道高斯分布的参数，这时定义高斯分布，然后用数据去学高斯分布的参数。

GAN的生成模型最后可以通过噪声生成一个完整的真实数据（比如人脸），说明生成模型掌握了从随机噪声到人脸数据的分布规律（GAN一开始并不知道规律是什么，通过一次次训练后学习到的真实样本集的数据分布）
